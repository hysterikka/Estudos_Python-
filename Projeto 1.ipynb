base de importacao

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

df = pd.read_csv("D:/Downloads/archive/earthquakes_2023_global.csv")
#exibir cabecalhos e os dados da tabela
df.head()
df.info()
```

Limpeza de dados

```python
#apagando os espacos vazios
df.dropna(inplace=True)
#estabelecendo a correlacao entre os dados da tabela
df.corr()
#valores Ãºnicos na tabela
df.nunique()
#exibindo a nova tabela
print(df)
```

Plotagens

Magnitudes X Tipos de magnitude

```python
#diagrama de dispersao
df.plot(kind = 'scatter', x='magType', y= 'mag', color = 'skyblue', edgecolor='white')
plt.show()
```

Magnitude (Histograma)

```python
#plotagem do histograma
plt.hist(df['mag'], bins = 4, color='skyblue', edgecolor='white')
#legendas
plt.xlabel('Valores')
plt.ylabel('Frequencia')
plt.show()
```

Mapa de dados coletados

```python
#figura  e eixos do mapa
plt.figure(figsize= (10 , 6))
ax=plt.axes(projection=ccrs.PlateCarree())
#desenho da costa
ax.coastlines()
#plotagem de pontos
ax.scatter(df['longitude'], df['latitude'], transform=ccrs.PlateCarree(), s=10, color='skyblue', marker='o', alpha=0.7)

#tÃ­tulo
plt.title('Mapa com pontos plotados')
#exibir mapa
plt.show()
```
